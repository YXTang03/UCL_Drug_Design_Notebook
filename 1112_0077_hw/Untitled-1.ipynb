```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import matplotlib
matplotlib.rcParams['font.family'] = 'Times New Roman'
```

```python
df = pd.DataFrame(
    {
        'mor': ['Morphine', 'Fentanyl', 'Methadone', 'DAMGO', 'Compound\n1', 'Compound\n2', 'Compound\n3', 'Compound\n4'],
        'gp_potency':[7.4, 9.48, 8.19, 8.73,6.6, 6.4, 7.8, 8.1],
        'gp_efficacy':[100, 106, 106, 101,96, 104, 95, 84], 
        'b_potency':[6.3, 6.9, 6.1, 6.42,5.2, 5.1, 6.6, 7.3], 
        'b_efficacy':[100, 313, 356, 434,55, 56, 15, 15]
    }
)
df
```

```python
fig, ax = plt.subplots(1, 2, figsize = (12, 5))
ax[0].plot(df['mor'], df['gp_potency'], marker = 'o', label = 'G-protein')
ax[0].plot(df['mor'], df['b_potency'], marker = 'o', label = r'$\beta$-arrestin')
ax[0].set_ylabel('Potency (pEC50)')

ax[1].plot(df['mor'], df['gp_efficacy'], marker = 'o', label = 'G-protein')
ax[1].plot(df['mor'], df['b_efficacy'], marker = 'o', label = r'$\beta$-arrestin')
ax[1].set_ylabel('Efficacy (%)')

for i in range(len(ax)):
    ax[i].legend()
    ax[i].tick_params(axis='x', labelrotation= 45)

plt.savefig('biased_vs_nonbiased.png', dpi = 600, bbox_inches='tight')
```
