```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sns
rcParams['font.family'] = 'serif'
rcParams['font.serif'] = ['Times New Roman']
```

```python

atp_conc = ['0', '0.03', '0.1', '0.3', '1', '3']
v_true = np.array([13.92199, 14.04078, 15.31843,	13.89617, 12.35871,	9.25853])
k_true = np.array([82.54831, 88.04961, 130.30483, 129.53799, 215.3569, 462.7184])
v_linear = np.array([12.2594091, 12.71294177, 11.79662616, 12.79754287, 11.31990038,6.618133686])
k_linear = np.array([53.29753586, 63.99389779, 60.83095435, 105.359355, 181.9314014,277.9775645])

fig, ax = plt.subplots(1, 2, figsize = (14, 5))
ax[0].spines['right'].set_visible(False)
ax[0].spines['top'].set_visible(False)
x = np.arange(len(atp_conc))
width = 0.35
alpha = 0.9
markersize = 3.5
v_color = sns.color_palette('viridis', 12)
k_color = sns.color_palette('plasma_r', 12)

rects1 = ax[0].bar(
    x-width/2, v_true, 
    width, 
    alpha = alpha, facecolor = v_color[0::2])
rects2 = ax[0].bar(
    x+width/2, v_linear, 
    width, 
    alpha = alpha, facecolor = v_color[1::2])

line_mm_v = ax[0].plot(
    x-width/2, v_true, color = 'black',
    marker = 'o', markersize = markersize,
    #label = 'Michaelis-Menten ($\mu mol/L$)'
)
line_lb_v = ax[0].plot(
    x+width/2, v_linear, color = 'dimgrey',
    marker = 'x', markersize = markersize,
    #label = r'Lineweaver-Burk ($\frac{\mu mol}{mg\cdot min}$)'
)
#ax[0].set_title('V$_{max}$')
for i in range(len(v_true)):
    ax[0].text((x-width)[i], (v_true+0.3)[i], f'{v_true[i]:.2f}')
    ax[0].text((x)[i], (v_linear-0.7)[i], f'{v_linear[i]:.2f}', color = 'black')


ax[1].spines['right'].set_visible(False)
ax[1].spines['top'].set_visible(False)
rects3 = ax[1].bar(
    x-width/2, k_true, 
    width, 
    alpha = alpha, facecolor = k_color[0::2])
rects4 = ax[1].bar(
    x+width/2, k_linear, 
    width, 
    alpha = alpha, facecolor = k_color[1::2])

line_mm_k = ax[1].plot(
    x-width/2, k_true, color = 'black',
    marker = 'o', markersize = markersize,
    label = 'Michaelis-Menten'
)
line_lb_k = ax[1].plot(
    x+width/2, k_linear, color = 'dimgrey',
    marker = 'x', markersize = markersize,
    label = 'Lineweaver-Burk'
)
#ax[1].set_title('K$_{m}$')
for i in range(len(v_true)):
    ax[1].text((x-width)[i], (k_true+15)[i], f'{k_true[i]:.2f}')
    ax[1].text((x+width*0.5)[i], (k_linear-20)[i], f'{k_linear[i]:.2f}', color = 'black')

units = ['V$_{max}$\n($\\frac{\mu mol}{mg\cdot min}$)','K$_{m}$\n($\mu mol/L$)']
for i in range(2):
    ax[i].set_xticks(x)
    ax[i].set_xticklabels(atp_conc)
    ax[i].set_ylabel(f'{units[i]}')
    ax[i].set_xlabel('ATP Concentration (mmol/L)')
    #ax[i].legend()
    ax[1].legend(bbox_to_anchor=(1, 1), loc='upper left', borderaxespad=0)
    #ax[i].legend(bbox_to_anchor=(0.5, -0.15), borderaxespad=0, ncol = 2)
plt.savefig('MM_vs_LB2.png', dpi = 600, bbox_inches='tight')
```
